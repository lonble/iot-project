```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("clean_data.csv")
```

```python
df['run_id'] = (
    df['environment'] + "_" +
    df['test'].astype(str) + "_" +
    df['sensor']
)
```

```python
def normalize(x):
    return (x - x.min())/(x.max() - x.min())
```

```python
def create_windows(signal, frame_length=500, overlap=0.5):

    step = int(frame_length*(1-overlap))

    windows = []

    for i in range(0,len(signal)-frame_length,step):
        windows.append(signal[i:i+frame_length])

    return np.array(windows)
```

```python
X=[]
y_env=[]
y_sensor=[]

for run_id,group in df.groupby("run_id"):

    rssi = group['rssi'].values

    diff = np.diff(rssi)

    norm = normalize(diff)

    windows = create_windows(norm,500,0.5)

    for w in windows:

        X.append(w)
        y_env.append(group['environment'].iloc[0])
        y_sensor.append(group['sensor'].iloc[0])

X = np.array(X)

print("Dataset shape:",X.shape)
```

```python
np.savez(
    "windows_500_50.npz",
    X=X,
    y_env=y_env,
    y_sensor=y_sensor
)

print("Dataset saved")
```
