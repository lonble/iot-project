```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_data.csv")
```

```python
sample = df[df['environment']=="garden"]

plt.figure(figsize=(12,4))
plt.plot(sample['rssi'].values[:2000])

plt.title("RSSI Time Series Example (Garden)")
plt.xlabel("Sample")
plt.ylabel("RSSI")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="environment", y="rssi", data=df)

plt.title("RSSI Distribution per Environment")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="sensor", y="rssi", data=df)

plt.title("RSSI Distribution per Sensor")

plt.show()
```

```python
results = pd.read_csv("experiment_results.csv")

results
```

```python
plt.figure(figsize=(7,5))

sns.barplot(
    x="Model",
    y="Accuracy",
    data=results
)

plt.title("CNN vs ResNet Performance")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.barplot(
    x="Frame",
    y="Accuracy",
    hue="Model",
    data=results
)

plt.title("Effect of Frame Size")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.barplot(
    x="Overlap",
    y="Accuracy",
    hue="Model",
    data=results
)

plt.title("Effect of Window Overlap")

plt.show()
```
